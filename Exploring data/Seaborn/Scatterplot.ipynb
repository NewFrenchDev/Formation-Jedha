```python
sns.scatterplot(data=df, x='LotArea', y='SalePrice')
```

```python
mask = 'LotArea < 20000 & SalePrice < 500000'
df_reduce_LotArea_SalePrice = df.query(mask)

sns.scatterplot(data=df_reduce_LotArea_SalePrice, x='LotArea', y='SalePrice')
```

```python
sns.relplot(x="LotFrontage", y="LotArea", data=df)
```
