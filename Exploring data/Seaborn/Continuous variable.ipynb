```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./sales_predictions.csv")
```

```python
df.head()
```

```python
df_sample = df.sample(25)
df_sample.head()
```

```python
ax = sns.relplot(data=df_sample, x='date', y='item_price', height=10, aspect=2, kind='line')
ax.set_xticklabels(fontsize=12, rotation=50)
```

```python
df_sample['date'] = pd.to_datetime(df_sample['date'])
df_sample['date']
```

```python
ax = sns.relplot(data=df_sample, x='date', y='item_price', height=10, aspect=2, kind='line')
ax.set_xticklabels(fontsize=12, rotation=50)
```
