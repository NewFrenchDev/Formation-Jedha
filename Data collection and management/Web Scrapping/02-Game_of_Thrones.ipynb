````python
import requests
import pandas as pd
import json
````

2. Go to the URL indicated in the description above. Using `requests`, get data on a character in the series. It doesn't matter who the character is.

````python
url = 'https://anapioficeandfire.com/'
response = requests.get(url)
response
````

````python
#Access to API characters
characters_access = 'api/characters/'
#Access to a specific character by his ID
characterID = '54'
#Get the information about the character
charac_selected = requests.get(url + characters_access + characterID)

charac_selected.content
````

3. Extract these data in JSON format:

````python
#Information extracted in json
charac_selected.json()
````

4. Now extract the JSON keys you have obtained and create a DataFrame that will have the JSON keys you extracted as a column.

**Hint**: we can use: `r.json().keys()`.

````python
#Get the names for the dataframe where we will store information for each personnage
column_names = charac_selected.json().keys()
display(column_names)

#Dataframe with columns = keys
df = pd.DataFrame(columns=column_names)
display(df)
````

5. Create a loop to extract data from the firsts 150 characters of the series and add them to your DataFrame.

````python
for i in range (1, 151):
    #Select the character by his ID
    characterID = i
    #Access to the caracter information
    charac_selected = requests.get(url + characters_access + str(characterID))  
    #Convert information to json keys = columns
    charac_info = charac_selected.json()
    #Add the information converted in the dataframe
    df = df.append(charac_info, ignore_index=True) 
    #Message to display when one personnage is done
    print(str(i) + " personnage done")
````

Let's display our dataframe filled by the information get through the API
