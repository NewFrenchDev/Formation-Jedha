```python
try_counter = 0
running = True

#First question
question1 = "Quelle est la capitale d'Allemagne?"
answer1 = ""
solution1  = "Berlin"

#Second question
question2 = "Quelle est la capitale de France?"
answer2 = ""
solution2  = "Paris"

#Third question
question3 = "Quelle est la capitale de Bulgarie?"
answer3 = ""
solution3  = "Sofia"
```

```python
while running:
    answer1 = input(question1)
    #Check format
    answer1_reformat = answer1[0].upper() + answer1[1:].lower()

    if answer1_reformat != solution1 and try_counter < 3:
        print("Too bad! that is not the correct answer")
        try_counter += 1
    elif answer1_reformat == solution1 and try_counter < 3:
        try_counter = 0
        print("Good job! This is the right answer")
        print("\n\n")
        break
    else: 
        print(":( you lost the game!")
        try_counter = 0
        running = False

while running:
    answer2 = input(question2)
    #Check format
    answer2_reformat = answer2[0].upper() + answer2[1:].lower()

    if answer2_reformat != solution2 and try_counter < 3:
        print("Too bad! that is not the correct answer")
        try_counter += 1
    elif answer2_reformat == solution2 and try_counter < 3:
        try_counter = 0
        print("Good job! This is the right answer")
        print("\n\n")
        break
    else: 
        print(":( you lost the game!")
        try_counter = 0
        running = False

        
while running:
    answer3 = input(question3)
    #Check format
    answer3_reformat = answer3[0].upper() + answer3[1:].lower()

    if answer3_reformat != solution3 and try_counter < 3:
        print("Too bad! that is not the correct answer")
        try_counter += 1
    elif answer3_reformat == solution3 and try_counter < 3:
        try_counter = 0
        print("Good job! This is the right answer")
        print("\n\n")
        break
    else: 
        print(":( you lost the game!")
        try_counter = 0
        running = False        
```
